```python
# Write a function to find actors who starred the most amount of times in top movies.
# (imdb rating 8 or more, metascore 90 or more)
# Who is at the top of the list?
task[8] = 5

def find_top_actors(df):
    top_movies = df[(df['rating'] >= 8) & (df['metascore'] >= 90)]
    actors = pd.concat([top_movies['actor1'], top_movies['actor2'], top_movies['actor3'], top_movies['actor4']])
    actor_counts = actors.value_counts()
    # Identify the top actor
    top_actor = actor_counts.idxmax()
    top_actor_count = actor_counts.max()

    return actor_counts, top_actor, top_actor_count

actor_counts, top_actor, top_actor_count = find_top_actors(imdb)

print("\nActor counts in top movies:\n", actor_counts)
print(f"\nTop actor: {top_actor} with {top_actor_count} appearances in top movies.")
```

```python
# Write a function which will find the most common genre by a decade.
# What is the most common genre of 1970s? Or 1980s? Why?
task[12] = 5
# Function to find the most common genre by decade
def most_common_genre_by_decade(df, decade):
    # Ensure the 'year' column is of integer type
    df['year'] = pd.to_numeric(df['year'], errors='coerce')

    # Filter the DataFrame for the specified decade
    start_year = decade
    end_year = start_year + 10
    decade_movies = df[(df['year'] >= start_year) & (df['year'] < end_year)]

    # Split genres and explode to count occurrences
    if not decade_movies.empty:
        all_genres = decade_movies['genre'].str.split(',').explode()
        genre_counts = all_genres.value_counts()

        # Get the most common genre
        most_common_genre = genre_counts.idxmax()
        return most_common_genre, genre_counts[most_common_genre]
    else:
        return None, 0
# Example usage for 1970s and 1980s
decades = [1970, 1980]
for decade in decades:
    most_common_genre, count = most_common_genre_by_decade(imdb, decade)
    if most_common_genre is not None:
        print(f"The most common genre in the {decade}s is '{most_common_genre}' with {count} occurrences.")
    else:
        print(f"No movies found for the {decade}s.")
```

```python
# Find great movies which are not well known.
# Get movie names which have 20,000 votes or less, while
# having rating of 8 or more. Make an option for metascore (90 or more).
task[13] = 5
import pandas as pd
def find_great_unknown_movies_v2(df, use_metascore=False):
    df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
    df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
    df['metascore'] = pd.to_numeric(df['metascore'], errors='coerce')
    query_string = "(votes <= 20000) & (rating >= 8)"

    if use_metascore:
        query_string += " & (metascore >= 90)"

    great_unknown_movies = df.query(query_string)

    if great_unknown_movies.empty:
        print("No great movies found that are not well known.")
        return pd.DataFrame(columns=['title', 'rating', 'metascore', 'votes'])

    return great_unknown_movies[['title', 'rating', 'metascore', 'votes']]

unknown_movies_v2 = find_great_unknown_movies_v2(imdb, use_metascore=True)
if not unknown_movies_v2.empty:
    print("Great movies that are not well known:\n", unknown_movies_v2)
else:
    print("Great movies that are not well known: None found.")
```

```python
# Write a function which will return a dictionary of statistics about a director:
# - Movie Title Series with year
# - Best rated movie(s)
# - Worst rated movie(s)
# - The average imdb rating of movies
# - The average metascore of movies
# - The average runtime of movies
# - The total gross (box office) revenue
# - Genres of movies directed
task[14] = 10
import pandas as pd

def director_statistics(df, director_name):
    director_movies = df[df['director'].str.lower() == director_name.lower()]

    if director_movies.empty:
        return {"Error": "No movies found for this director."}

    movie_titles = director_movies[['title', 'year']].set_index('year').to_dict()['title']
    best_rated_movies = director_movies[director_movies['rating'] == director_movies['rating'].max()]['title'].tolist()
    worst_rated_movies = director_movies[director_movies['rating'] == director_movies['rating'].min()]['title'].tolist()
    average_rating = director_movies['rating'].mean()
    average_metascore = director_movies['metascore'].mean() if 'metascore' in director_movies.columns else None
    average_runtime = director_movies['runtime'].mean() if 'runtime' in director_movies.columns else None
    total_gross = director_movies['gross'].sum() if 'gross' in director_movies.columns else 0  # Assuming 'gross' column exists
    genres = director_movies['genre'].unique().tolist() if 'genre' in director_movies.columns else []

    stats = {
        "Movie Titles": movie_titles,
        "Best Rated Movie(s)": best_rated_movies,
        "Worst Rated Movie(s)": worst_rated_movies,
        "Average IMDb Rating": average_rating,
        "Average Metascore": average_metascore,
        "Average Runtime": average_runtime,
        "Total Gross Revenue": total_gross,
        "Genres": genres
    }

    return stats

# Example usage
director_name = "Christopher Nolan"  # It can be any director name, I'll try with this
director_stats = director_statistics(imdb, director_name)

# Print out the statistics
for key, value in director_stats.items():
    print(f"{key}: {value}")
```

```python
# Write a function which should plot a histogram showing movie frequency by every year
# Function should accept a range of start year and end year.
task[15] = 10
import pandas as pd
import matplotlib.pyplot as plt

def plot_movie_frequency_by_year(df, start_year, end_year):
    filtered_movies = df[(df['year'] >= start_year) & (df['year'] <= end_year)]

    movie_counts = filtered_movies['year'].value_counts().sort_index()

    plt.figure(figsize=(12, 6))
    plt.bar(movie_counts.index, movie_counts.values, color='skyblue')
    plt.title(f'Movie Frequency by Year from {start_year} to {end_year}', fontsize=16)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Number of Movies', fontsize=14)
    plt.xticks(movie_counts.index, rotation=90)
    plt.grid(axis='y', linestyle='--')
    plt.tight_layout()
    plt.show()

# Example usage
start_year = 1920
end_year = 2023
plot_movie_frequency_by_year(imdb, start_year, end_year)
```
